The metric being measured here is safety based on the amount of playgrounds present in a neighborhood. The more playgrounds means on average there's more kids nearby. If a certain neighborhood has a lot of children it means it's generally more safe. Neighborhoods with more playgrounds seem more appealing to families, especially those with many kids.

```python
import pandas as pd
import numpy as np
import geopandas as gdp
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Read data from playground csv file
playground = pd.read_csv("playground.csv")
```

```python
playground[['neighborhood', 'name']]
```

```python
a = playground['neighborhood'].value_counts()
print(a)
```

```python
a.plot(x = 'neighborhood', y = 'name',kind = 'bar', figsize = (20,20), )
```

According to the graphs shown above, it's obvious Squirrel Hill South is the best and most fun neighborhood for children to live in, because it has the most playgrounds.
